```python
def read_data(filename):
    with open(filename, "r") as f:
        data = f.readlines()
    return data
```

```python
def datatype_conversion(line):   
    lsplit = line.split(",")
    lsplit[5] = float(line.split(",")[5].replace("\"","").replace("SGD",""))
    lsplit[16] = float(line.split(",")[5].replace("\"","").replace("SGD",""))
    line = (','.join(str(l) for l in lsplit))
    return line
```

```python
def segregate_data(data):
    finedata = []
    finedata.append(data[0])
    anamolydata = []
    anamolydata.append(data[0])
    for line in data[1:]:
        linesplit = line.split(",")
        if len(linesplit) == 25:
            finedata.append(datatype_conversion(line))
        else:
            anamolydata.append(line)
    return finedata, anamolydata  
```

```python
def write_to_file(filename, data):
    with open(filename, "w+") as f:
        for line in data:
            f.write(line)
        f.close()
```

```python
data = read_data(original_file)
finedata, anamolydata = segregate_data(data)
write_to_file(fine_data_file, finedata)
write_to_file(anamoly_data_file, anamolydata)
```
